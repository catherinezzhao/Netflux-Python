```runSimulation()``` ran things with the default parameters in the .xlsx file. To follow the tutorial, we will add another line below this one.

runSimulation accepts 4 arguments: ```ID```, ```param_type```, ```value```, ```Multiple```. We will first talk about the first three arguments.

1. ```ID```: 
This is the species ID or reaction ID --> e.g., "A" for species ID, "r2" for reaction ID

2. ```param_type```: 
This can be tau, Ymax, Yinit, Weight, n, or EC50    --> e.g., "Weight"

3. ```value```: 
This is the value you want to change it to                --> e.g., "1"

**To change the reaction weight for “r2: => A” to 1:**

````python
#---ONLY MANIPULATE STUFF BETWEEN THESE LINES--------
runSimulation()
runSimulation("r2","Weight","1")

#----------------------------------------------------
````

### Step 6:
Simulate simultaneous stimulation with both A and B by changing their input reaction weights to 1.

This is slightly different from the other runs. 

````python
#---ONLY MANIPULATE STUFF BETWEEN THESE LINES--------
runSimulation()
runSimulation("r2","Weight","1")
runSimulation("r2","Weight","0")
runSimulation("r1","Weight","1")
runSimulation("r1","Weight","0")
runSimulation("r2","Weight","1", True)
runSimulation()

#----------------------------------------------------
````

We specified that reaction r2 will have a weight of 1, but we need to also handle reaction r1. 

To do this, we set ```Multiple```, the 4th argument, to be **True**. By doing so, we're saying that we want to simulate multiple parameters at once. 

When we run ```_run.py```, we will follow the prompts in our IDE console.
* We will enter "r", since r1 is a reaction ID
* We will enter "r1" for the reaction ID
* We will enter "Weight" to change r1's Weight
* We will enter "1" to change the value to 1
* Finally, we will enter "n" since we have no more parameters to change for this run. If you wanted to change more than 2 parameters for a run, you would enter "y" and continue.
